```python
import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd

# Get Data: this ex will only use part of it (i.e. rows 750-1500)
df = pd.read_csv('/Users/jun/Desktop/503 project2/cancer_us.csv')


trace1 = go.Scatter3d(
    x=df['YEAR'],
    y=df['Mortality_COUNT'],
    z=df['Incidence_COUNT'],
    text=df['Mortality_AGE_ADJUSTED_RATE'],
    mode='markers',
    marker=dict(
        sizemode='diameter',
        sizeref=10,
        size=df['Mortality_AGE_ADJUSTED_RATE'],
        color = df['Mortality_AGE_ADJUSTED_RATE'],
        colorscale = 'Hot',
        colorbar = dict(title = 'Incidence<br>Age Adjusted Rate'),
        line=dict(color='rgb(140, 140, 170)')
    )
)


trace2 = go.Scatter3d(
    x=df['YEAR'],
    y=df['Mortality_COUNT'],
    z=df['Incidence_COUNT'],
    text=df['Incidence_AGE_ADJUSTED_RATE'],
    mode='markers',
    marker=dict(
        sizemode='diameter',
        sizeref=10,
        size=df['Incidence_AGE_ADJUSTED_RATE'],
        color = df['Incidence_AGE_ADJUSTED_RATE'],
        colorscale = 'Viridis',
        colorbar = dict(title = 'Mortality<br>Age Adjusted Rate'),
        line=dict(color='rgb(140, 140, 170)')
    )
)



data=[trace1, trace2]

#layout=go.Layout(height=800, width=800, title='Examining Population and Life Expectancy Over Time')

layout = go.Layout(
                    scene = dict(
                        xaxis = dict(
                            title='YEAR'),
                        yaxis = dict(
                            title='Mortality_COUNT'),
                        zaxis = dict(
                            title='Incidence_COUNT'),),
                    width=700,
                    height=800,
                    title="Cancer Mortality vs. Cancer Incidence",
                    showlegend=False
                  )



fig=go.Figure(data=data, layout=layout)
py.iplot(fig, filename='3DBubble')
```

```python
steps = []
for i in range(len(df)):
    step = dict(
        method = 'restyle',  
        args = ['visible', [False] * len(df)],
    )
    step['args'][1][i] = True # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [dict(
    active = 10,
    currentvalue = {"prefix": "Frequency: "},
    pad = {"t": 50},
    steps = steps
)]

layout = dict(sliders=sliders)
```
